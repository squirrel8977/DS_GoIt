```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

Прочитайте дані за допомогою методу read_html з таблиці "Коефіцієнт народжуваності в регіонах України (1950—2019)".

```python
url = 'https://uk.wikipedia.org/wiki/%D0%9D%D0%B0%D1%81%D0%B5%D0%BB%D0%B5%D0%BD%D0%BD%D1%8F_%D0%A3%D0%BA%D1%80%D0%B0%D1%97%D0%BD%D0%B8'
df = pd.read_html(url, match='Коефіцієнт народжуваності в регіонах України', thousands=".", decimal=",")[0]
```

Вивести перші рядки таблиці за допомогою методу head.

```python
df.head()
```

Визначте кількість рядків та стовпців у датафреймі (атрибут shape).

```python
df.shape
```

Замініть у таблиці значення "—" на значення NaN.

```python
df[["2014", "2019"]] = df[["2014", "2019"]].apply(pd.to_numeric, errors="coerce")
df.dtypes
```

Порахуйте, яка частка пропусків міститься в кожній колонці (використовуйте методи isnull та sum).

```python
df.isnull().sum() / len(df)
```

Видаліть з таблиці дані по всій країні, останній рядок таблиці.

```python
df.drop(df.index[-1], axis=0, inplace=True)
df.tail()
```

Замініть відсутні дані в стовпцях середніми значеннями цих стовпців (метод fillna).

```python
df = df.fillna(df.mean(numeric_only=True), axis=0)
df.head()
```

Отримайте список регіонів, де рівень народжуваності у 2019 році був вищим за середній по Україні.

```python
avg = df['2019'].mean()
regions_above_avg_2019 = df[df['2019'] > avg]
regions_above_avg_2019
```

У якому регіоні була найвища народжуваність у 2014 році?

```python
highest_birth_2014 = df.loc[df['2014'].idxmax(), 'Регіон']
highest_birth_2014
```

Побудуйте стовпчикову діаграму народжуваності по регіонах у 2019 році.

```python
df.plot(kind='bar', x='Регіон', y='2019', legend=False)
plt.ylabel('Коефіцієнт народжуваності у 2019 році')
plt.title('Народжуваність по регіонах України у 2019 році')
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(df.columns[1:], df.iloc[-1][1:], color="teal", marker="D")
plt.xlabel("Рік")
plt.ylabel("Коефіцієнт народжуваності")
plt.grid()
plt.title("Динаміка коефіцієнта народжуваності - вся Україна (avg)")
plt.show()
```

```python
df.plot(kind='line', figsize=(12, 6))
plt.title('Динаміка народжуваності')
plt.xlabel('Рік')
plt.ylabel('Коефіцієнт народжуваності')
plt.grid(True)
plt.show()
```

```python
sns.boxplot(data=df)
plt.title('Розподіл народжуваності за роками')
plt.xlabel('Рік')
plt.ylabel('Коефіцієнт народжуваності')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(data=df.drop(['Регіон'], axis=1))
plt.xlabel('Рік')
plt.ylabel('Коефіцієнт народжуваності')
plt.title('Розподіл коефіцієнта народжуваності в регіонах за роками')
plt.show()
```
